````python
import zipfile
import os

# Define the paths to the zip files
zip_train_path = r"/content/train-20260411T114838Z-3-001.zip"
zip_valid_path = r"/content/valid-20260411T114844Z-3-001.zip"
zip_test_path  = r"/content/test-20260411T114820Z-3-001.zip"

# Define target directories for unzipping
output_train_dir = "/content/train"
output_valid_dir = "/content/valid"
output_test_dir  = "/content/test"

# Create directories if they don't exist
os.makedirs(output_train_dir, exist_ok=True)
os.makedirs(output_valid_dir, exist_ok=True)
os.makedirs(output_test_dir, exist_ok=True)

# Unzip the datasets
with zipfile.ZipFile(zip_train_path, 'r') as zip_ref:
    zip_ref.extractall(output_train_dir)

with zipfile.ZipFile(zip_valid_path, 'r') as zip_ref:
    zip_ref.extractall(output_valid_dir)

with zipfile.ZipFile(zip_test_path, 'r') as zip_ref:
    zip_ref.extractall(output_test_dir)

# Update the paths to point to the unzipped directories
train_path = output_train_dir
valid_path = output_valid_dir
test_path  = output_test_dir

from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ImageDataGenerator with Data Augmentation for Training
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,         # rotate images
    width_shift_range=0.1,     # horizontal shift
    height_shift_range=0.1,    # vertical shift
    zoom_range=0.2,            # zoom in/out
    horizontal_flip=True,      # flip images
    brightness_range=[0.8,1.2] # brightness change
)

train_data = train_gen.flow_from_directory(
    train_path,
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical"
)

# ImageDataGenerator without Data Augmentation for Validation and Test
valid_gen = ImageDataGenerator(rescale=1./255)
test_gen = ImageDataGenerator(rescale=1./255)

valid_data = valid_gen.flow_from_directory(
    valid_path,
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical"
)

test_data = test_gen.flow_from_directory(
    test_path,
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical"
)
````

````python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import ResNet50, MobileNetV2, EfficientNetB0

# Define input shape matching your ImageDataGenerator target_size
IMG_SHAPE = (224, 224, 3)

# --- ResNet50 Example ---

# 1. Load the pre-trained ResNet50 model without its top classification layer
base_model_resnet = ResNet50(
    input_shape=IMG_SHAPE,
    include_top=False, # Do not include the ImageNet classifier at the top
    weights='imagenet' # Load weights pre-trained on ImageNet
)

# 2. Freeze the base model layers so they are not updated during training
base_model_resnet.trainable = False

# 3. Create new custom top layers for classification
#    The number of classes comes from your ImageDataGenerator (train_data.num_classes)
num_classes = train_data.num_classes

inputs = tf.keras.Input(shape=IMG_SHAPE)
x = base_model_resnet(inputs, training=False) # Run the base model in inference mode
x = layers.GlobalAveragePooling2D()(x) # Reduce spatial dimensions
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)

model_resnet_tl = tf.keras.Model(inputs, outputs, name='ResNet50_TransferLearning')

# 4. Compile the model
model_resnet_tl.compile(optimizer='adam',
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])

# 5. Display the model summary
model_resnet_tl.summary()
````

````python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
from tensorflow.keras import backend as K # Added this import

# --- Evaluation Function ---
def evaluate_model_and_plot(model_name, model, history, test_data):
    print(f"\n--- Evaluating {model_name} ---")

    # 1. Evaluate on Test Data
    # It's crucial that the model is compiled with the same metrics (Precision, Recall)
    # for model.evaluate to return them.
    results = model.evaluate(test_data, verbose=0)
    # Assuming results order matches common_metrics: loss, accuracy, precision, recall
    test_loss, test_accuracy, test_precision, test_recall = results[0], results[1], results[2], results[3]


    print(f"{model_name} - Test Loss: {test_loss:.4f}")
    print(f"{model_name} - Test Accuracy: {test_accuracy:.4f}")

    # Calculate F1-score for test set
    test_f1 = 2 * ((test_precision * test_recall) / (test_precision + test_recall + K.epsilon()))
    print(f"{model_name} - Test Precision: {test_precision:.4f}")
    print(f"{model_name} - Test Recall: {test_recall:.4f}")
    print(f"{model_name} - Test F1-Score: {test_f1:.4f}")

    # 2. Get predictions and true labels for Confusion Matrix and Classification Report
    # Reset test_data generator to ensure order
    test_data.reset()
    # Predict the probabilities for each class
    predictions = model.predict(test_data)
    # Get the class with the highest probability
    predicted_classes = np.argmax(predictions, axis=1)

    # Get true labels
    true_classes = test_data.classes
    class_labels = list(test_data.class_indices.keys())

    # 3. Confusion Matrix
    print(f"\n--- Confusion Matrix for {model_name} ---")
    cm = confusion_matrix(true_classes, predicted_classes)
    print(cm)

    # Plotting Confusion Matrix
    plt.figure(figsize=(8, 6))
    import seaborn as sns
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # 4. Classification Report
    print(f"\n--- Classification Report for {model_name} ---")
    print(classification_report(true_classes, predicted_classes, target_names=class_labels))

    # 5. Plot Accuracy and Loss Curves
    print(f"\n--- Plotting Learning Curves for {model_name} ---")
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title(f'{model_name} - Training and Validation Accuracy')

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title(f'{model_name} - Training and Validation Loss')
    plt.show()

# --- Evaluate Custom CNN Model ---
# Ensure `model` and `history_cnn` exist from previous steps
evaluate_model_and_plot('Custom CNN Model', model, history_cnn, test_data)

# --- Evaluate ResNet50 Transfer Learning Model ---
# Ensure `model_resnet_tl` and `history_resnet_tl` exist from previous steps
evaluate_model_and_plot('ResNet50 Transfer Learning Model', model_resnet_tl, history_resnet_tl, test_data)
````

````python
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, Callback
from tensorflow.keras.metrics import Precision, Recall
from tensorflow.keras import backend as K
import numpy as np
import os

# --- Custom F1-score Callback ---
# This callback calculates F1-score from reported Precision and Recall at the end of each epoch.
class F1ScoreMetric(Callback):
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        val_precision = logs.get('val_precision')
        val_recall = logs.get('val_recall')

        if val_precision is not None and val_recall is not None and (val_precision + val_recall) > K.epsilon():
            _val_f1 = 2 * ((val_precision * val_recall) / (val_precision + val_recall + K.epsilon()))
            logs['val_f1'] = _val_f1
            print(f" - val_f1: {_val_f1:.4f}")
        else:
            # Handle cases where precision or recall might be zero, leading to division by zero
            logs['val_f1'] = 0.0 # Or np.nan for 'Not a Number'
            print(" - val_f1: N/A (Precision or Recall not available or division by zero)")

# --- Callbacks setup ---
epochs = 1 # You can adjust the number of epochs

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10, # Stop if val_loss doesn't improve for 10 epochs
    restore_best_weights=True # Restore model weights from the epoch with the best value of the monitored quantity.
)

# ModelCheckpoint for the custom CNN model
checkpoint_cnn = ModelCheckpoint(
    filepath='best_custom_cnn_model.keras',
    monitor='val_accuracy', # Monitor validation accuracy
    save_best_only=True,   # Save only the best model
    mode='max',            # Higher accuracy is better
    verbose=1
)

# ModelCheckpoint for the transfer learning model
checkpoint_resnet_tl = ModelCheckpoint(
    filepath='best_resnet_tl_model.keras',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

# Combine all callbacks
callbacks_cnn = [early_stopping, checkpoint_cnn, F1ScoreMetric()]
callbacks_resnet_tl = [early_stopping, checkpoint_resnet_tl, F1ScoreMetric()]

# --- Common metrics for compilation ---
common_metrics = [
    'accuracy',
    Precision(name='precision'),
    Recall(name='recall')
]

# --- Training Custom CNN Model ---
print("\n--- Training Custom CNN Model ---")
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=common_metrics
)

history_cnn = model.fit(
    train_data,
    epochs=epochs,
    validation_data=valid_data,
    callbacks=callbacks_cnn
)

# --- Training ResNet50 Transfer Learning Model ---
print("\n--- Training ResNet50 Transfer Learning Model ---")
# Ensure the model is compiled again, especially if metrics change or it's a new training run.
model_resnet_tl.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=common_metrics
)

history_resnet_tl = model_resnet_tl.fit(
    train_data,
    epochs=epochs,
    validation_data=valid_data,
    callbacks=callbacks_resnet_tl
)

print("\nTraining complete for both models.")
````

````python
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the number of classes from your ImageDataGenerator output
# Assuming train_data.num_classes correctly reflects the number of classes
num_classes = train_data.num_classes

model = models.Sequential([
    # First Convolutional Block
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Second Convolutional Block
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Third Convolutional Block
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Flatten the output for the dense layers
    layers.Flatten(),

    # Dense Layers
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    # Output Layer
    layers.Dense(num_classes, activation='softmax')
])

# Display the model summary
model.summary()
````
